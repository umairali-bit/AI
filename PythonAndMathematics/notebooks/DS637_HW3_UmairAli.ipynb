```python
import numpy as np
from functools import reduce
from time import perf_counter



```

```python
def myPi_numpy(n): return 4*np.sum(((-1)**(np.arange(1, n+1)+1))/(2*np.arange(1, n+1)-1))

#last week's
def myPi_loop(n): return 4*sum(((-1)**(k+1))/(2*k-1) for k in range(1, n+1))


#Quick tests + timing:
for n in [10, 100, 10_000]:
    print(n, myPi_numpy(n), myPi_loop(n))

n = 2_000_000
t0 = perf_counter(); a = myPi_loop(n);  t1 = perf_counter()
t2 = perf_counter(); b = myPi_numpy(n); t3 = perf_counter()
print("loop:", t1-t0, "numpy:", t3-t2, "diff:", abs(a-b))
```

```python
def manhattan(p, q): return sum(abs(a-b) for a, b in zip(p, q))

print(manhattan([1,2,3], [4,0,3]))   # 5
print(manhattan([0,0], [0,0]))      # 0
```

```python
def correlation(x, y): return float(np.corrcoef(np.array(x, dtype=float), np.array(y, dtype=float))[0,1])

print(correlation([1,2,3], [1,2,3]))   # close to 1.0
print(correlation([1,2,3], [3,2,1]))   # close to -1.0
```

```python
def fib_first_n(n): return [] if n<=0 else [0] if n==1 else reduce(lambda s,_: s+[s[-1]+s[-2]], range(n-2), [0,1])

print(fib_first_n(0))   # []
print(fib_first_n(1))   # [0]
print(fib_first_n(8))   # [0, 1, 1, 2, 3, 5, 8, 13]
```
