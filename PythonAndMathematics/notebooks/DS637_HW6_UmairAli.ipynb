# DS637 – Homework: Visualize Daily Treasury Yield Curve Rates


## 1. Load data and set Date as index

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Load the CSV file
df = pd.read_csv("treasury_yield_curve.csv")

# Convert Date column to datetime and set as index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

df.head()
```

## 2. Plot curves for all columns in one graph

```python
plt.figure(figsize=(10,6))

for col in df.columns:
    plt.plot(df.index, df[col], label=col)

plt.title("Daily Treasury Yield Curve Rates")
plt.xlabel("Date")
plt.ylabel("Yield (%)")
plt.legend()
plt.show()
```

## 3. Pairplot for short rates (less than one year)

```python
short_rates = df[['1 Mo','2 Mo','3 Mo','6 Mo']]

sns.pairplot(short_rates)
plt.show()
```

## 4. Scatter plot with regression line for 2 Mo and 3 Mo

```python
sns.regplot(x=df['2 Mo'], y=df['3 Mo'])
plt.xlabel("2 Month Rate")
plt.ylabel("3 Month Rate")
plt.title("Regression: 2 Mo vs 3 Mo")
plt.show()
```

## 5. KDE jointplot for 2 Mo and 3 Mo rates

```python
sns.jointplot(x='2 Mo', y='3 Mo', data=df, kind='kde')
plt.show()
```

## 6. Boxplot of all columns

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplot of Treasury Yield Rates")
plt.show()
```
