```python
def is_prime(x):
    if x <= 1:
        return False
    for i in range(2, int(x** 0.5) + 1):
        if x % i == 0:
            return False
    return True

# Test cases
print(is_prime(2))    # True
print(is_prime(7))    # True
print(is_prime(10))   # False
print(is_prime(1))    # False
    
```

```python
def sum_to_x(x):
    return sum(range(1, x + 1))

# Example tests
print(sum_to_x(5))
print(sum_to_x(10))
```
