```python
def squareRoot(a, tolerance=1e-10):

    x = a

    while True:
        next_x = 0.5 * (x + a / x)

        if abs(next_x - x) < tolerance:
            return next_x
            
        x = next_x

print(squareRoot(2))
print(squareRoot(9))
print(squareRoot(16))
       

```

```python
def myPi(n):
    s = 0.0
    for k in range(1, n + 1):
        term = ((-1) ** (k + 1)) / (2 * k - 1)
        s += term
    return 4 * s

print(myPi(1))
print(myPi(2))
print(myPi(10))
print(myPi(1000))
```

```python
def primes_up_to_1000():
    limit = 1000
    is_prime = [True] * (limit + 1)

    # 0 and 1 are not prime
    is_prime[0] = False
    is_prime[1] = False

    p = 2
    while p * p <= limit:  # same as p <= sqrt(limit)
        if is_prime[p]:
            # mark multiples of p as composite
            for multiple in range(p * p, limit + 1, p):
                is_prime[multiple] = False
        p += 1

    # return list of primes
    return [i for i in range(limit + 1) if is_prime[i]]

primes = primes_up_to_1000()
print(primes)
print("Count:", len(primes))

# Why range(p*p, ...) and not range(2*p, ...)?

#Because anything smaller than p*p (like 2p, 3p, 4p...) would already have been marked by smaller primes earlier.
```
